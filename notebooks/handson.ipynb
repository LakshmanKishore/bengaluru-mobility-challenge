```python
import cv2

# Define the input and output video paths
input_video_path = f'{BASE_PATH}/videos/Devasandra_Sgnl_JN_FIX_1_time_2024-05-14T07_30_02_000.mp4'
output_video_path = f'{BASE_PATH}/processed_videos/output_video_v2.mp4'
new_fps = 10  # Desired fps
scale_factor = 0.5  # Scale factor for reducing resolution

# Open the input video
cap = cv2.VideoCapture(input_video_path)

# Get the properties of the input video
original_fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH) * scale_factor)
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT) * scale_factor)
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Calculate the frame interval to match the new fps
frame_interval = int(original_fps / new_fps)

# Define the codec and create the VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Use 'XVID' for .avi format
out = cv2.VideoWriter(output_video_path, fourcc, new_fps, (width, height))

# Loop through the video frames and write every nth frame to the output video
frame_index = 0
success, frame = cap.read()

while success:
    if frame_index % frame_interval == 0:
        # Resize the frame
        frame = cv2.resize(frame, (width, height))
        out.write(frame)
    frame_index += 1
    success, frame = cap.read()

# Release the resources
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Video has been saved to {output_video_path} with {new_fps} fps and reduced resolution.")
```
